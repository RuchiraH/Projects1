### Spliting the data

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
```

### 1)Predicting the data using Logistics Regression

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression() 
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report
ac = accuracy_score(ytest,ypred)
cr = classification_report(ytest,ypred)
print("Accuracy score : ",ac)
print(cr)
```

### We have achieved an Average Accuracy of 90 % . Lets see if we can increase this accuracy by hyper tuning

### Using HTP(To increase accuracy we use HTP)

### 1)solver = "liblinear"

```python
logreg = LogisticRegression(solver = "liblinear")
logreg.fit(xtrain,ytrain)
ypred = logreg.predict(xtest)
```

```python
ac = accuracy_score(ytest,ypred)
cr = classification_report(ytest,ypred)
print("Accuracy score : ",ac)
print(cr)
```

### By using liblinear we get Accuracy of 90%

### 2)Predicting the data using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
dt = DecisionTreeClassifier()
```

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred = model.predict(xtest)
    print(accuracy_score(ytest,ypred))
    print(classification_report(ytest,ypred))
    return model
```

```python
mymodel(dt)
```

### By using Decision Tree we get average accuracy of 89% which is good but lets check whether we get more accurcy by hyper tunning

### Ploting decision tree

```python
fig,ax=plt.subplots(figsize=(25,10))
chart=tree.plot_tree(dt,feature_names=x.columns,filled=True,fontsize=10)
plt.savefig('Tree1.jpg')
```

```python
train = dt.score(xtrain,ytrain)
test = dt.score(xtest,ytest)
print(f"train score :{train} \ntest score : {test}")
```

```python
# as training accuracy is  high then training errors will be low 
# as training accuracy is  low then training errors will be high 
# low bias & High variance = overfitting 
```

### max_depth Parameter

```python
for i in range(1,50):
    dt1=DecisionTreeClassifier(max_depth=i)
    dt1.fit(xtrain,ytrain)
    ypred=dt1.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f' max depth: {i}, Accuracy: {ac}')
```

```python
dt2=DecisionTreeClassifier(max_depth=5)#remake the object everytime
mymodel(dt2)
```

### By using Decision Tree Max Depth function we achive the accuracy of 90%

```python
train = dt2.score(xtrain,ytrain)
test = dt2.score(xtest,ytest)
print(f"train score :{train} \ntest score : {test}")
```

### min_sample_split

```python
for i in range(50,100):
    dt3=DecisionTreeClassifier(min_samples_split=i)
    dt3.fit(xtrain,ytrain)
    ypred=dt3.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f' max depth: {i}, Accuracy: {ac}')
```

```python
dt4=DecisionTreeClassifier(min_samples_split=54)#remake the object everytime
mymodel(dt4)
```

### By using Decision Tree Min Sample Split function we achive the accuracy of 90%

```python
train = dt4.score(xtrain,ytrain)
test = dt4.score(xtest,ytest)
print(f"train score :{train} \ntest score : {test}")
```

### min_sample_leaf

```python
for i in range(1,50):
    dt5=DecisionTreeClassifier(min_samples_leaf=i)
    dt5.fit(xtrain,ytrain)
    ypred=dt5.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f' max depth: {i}, Accuracy: {ac}')
```

```python
dt6=DecisionTreeClassifier(min_samples_split=28)#remake the object everytime
mymodel(dt6)
```

### By using Decision Tree min_samples_leaf function we achive the accuracy of 90%

```python
train = dt6.score(xtrain,ytrain)
test = dt6.score(xtest,ytest)
print(f"train score :{train} \ntest score : {test}")
```

### Votting Classifier

```python
models =[]
models.append(("lr",LogisticRegression()))
models.append(("dt",DecisionTreeClassifier()))
```

```python
from sklearn.ensemble import VotingClassifier
vc = VotingClassifier(estimators = models)	# estimators --> model name
vc.fit(xtrain,ytrain)
ypred = vc.predict(xtest)
print(classification_report(ytest,ypred))
```

### By using Voting Classifier we get 90% of accuracy.

```python
from sklearn.ensemble import VotingClassifier
vc = VotingClassifier(estimators = models,voting='soft')
vc.fit(xtrain,ytrain)
ypred = vc.predict(xtest)
print(classification_report(ytest,ypred))
```

### By using Voting Classifier using soft voting we get 89% of accuracy.
