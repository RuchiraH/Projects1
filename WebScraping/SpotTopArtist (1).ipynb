Importing various libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

Web Scraping on Most streamed artists on Spotify

1. Here first we are importing libraries such as requests were the web browser sends request  to the http server and  it sends the message containing a status line.

2. We import the pandas library and give the alias as pd which is an open source library in Python.  It provides ready to use high-performance data structures(i.e. This means that pandas can operate on large datasets with minimal overhead.)and data analysis tools.

3. We import the BeautifulSoup library its a tool we will be using in our project to scrape the website

Get HTML from a website


```python
url='https://chartmasters.org/most-streamed-artists-ever-on-spotify/'
```

We have provided the url of the website 

```python
page=requests.get(url)
page
```

We had sent the request to the webpage and we have received the status line as response 200 that means we can go ahead with the webscrapping.

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

Here we used the html parser as the page is in the text format or the paragraph format and then we prettify it it is just to add the structure to the data

### Scrapping the Page

soup.find will find table tag and id related to required table

```python
table=soup.find("table", id="table_1")
```

Now that content is present in table and we only required the table body that is in tbody tag so we use find tbody

```python
table=table.find('tbody')
```

```python
table
```

```python
rows=table.find_all('tr')
```

```python
rows
```

```python
information=[]
for i in rows:
    cols=i.find_all('td')
    #print(cols)
    info=[c.text for c in cols]
    print(info)
    information.append(info)
print(information)
```

Here,we take an empty list and then we take for loop in rows which contains all tr tags, for every tr tag we find all td tags also and then we append the list which we took as an empty list

```python
Number=[]
Artist=[]
Lead_Streams=[]
Tracks=[]
One_Billion=[]
Hundred_Million=[]
Ten_Million=[]
One_Million=[]
Feat_Streams=[]
for i in information:
    Number.append(i[0])
    Artist.append(i[2])
    Lead_Streams.append(i[3])
    Tracks.append(i[4])
    One_Billion.append(i[5])
    Hundred_Million.append(i[6])
    Ten_Million.append(i[7])
    One_Million.append(i[8])
    Feat_Streams.append(i[9])
```

Creating a dataframe using pandas library

```python
Data={'Number':Number,'Artist':Artist,'Lead_Streams':Lead_Streams,'Tracks':Tracks,'One_Billion':One_Billion,'Hundred_Million':Hundred_Million,'Ten_Million':Ten_Million,'One_Million':One_Million,'Feat_Streams':Feat_Streams}
```

```python
Most_Streamed_Artists=pd.DataFrame(Data)
Most_Streamed_Artists
```

Saving data in CSV file.

```python
Most_Streamed_Artists.to_csv('Most Streamed Artists on Spotify.csv')
```
